```python
import pandas as pd
 
# ── Configuration ──────────────────────────────────────────────
INPUT_FILE  = "pcbi.1002321.s005.csv"
OUTPUT_FILE = "food_web_filtered.csv"
 
# Add any species codes you want to remove here
SPECIES_TO_REMOVE = {
    "CANAUR",
    "DAMKOR",
    "GAZGRA",
    "GAZTHO",
    "HETBRU",
    "HIPAMP",
    "KOBELL",
    "LYCPIC",
    "PAPANU",
    "PEDCAP",
    "REDRED",
    "RHAPUM",
    "OUROUR",
    "SYNCAF",
    "TRASCR"
}
# ──────────────────────────────────────────────────────────────
 
df = pd.read_csv(INPUT_FILE)
 
print(f"Rows before: {len(df)}")
 
# Remove any row where the species appears as predator OR prey
mask = df["pred_code"].isin(SPECIES_TO_REMOVE) | df["prey_code"].isin(SPECIES_TO_REMOVE)
removed = df[mask]
df_filtered = df[~mask]
 
print(f"Rows removed: {len(removed)}")
print(f"Rows after:   {len(df_filtered)}")
 
if len(removed) > 0:
    print("\nRemoved links:")
    print(removed.to_string(index=False))
 
df_filtered.to_csv(OUTPUT_FILE, index=False)
print(f"\nSaved to {OUTPUT_FILE}")
 




```
